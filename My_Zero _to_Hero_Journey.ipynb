# March, 18,2026, Calculator




```python
import math

print("----- Calculator -----")
print("Choose an operation:")
print("add, sub, mul, div, square, sqrt, log")
print("------------------------------------")

operation = input("Enter operation: ").lower()

# Addition
if operation == "add":
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    print("Result:", a + b)

# Subtraction
elif operation == "sub":
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    print("Result:", a - b)

# Multiplication
elif operation == "mul":
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    print("Result:", a * b)

# Division
elif operation == "div":
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    if b == 0:
        print("Error: Cannot divide by zero.")
    else:
        print("Result:", a / b)

# Square
elif operation == "square":
    a = float(input("Enter number: "))
    print("Result:", a ** 2)

# Square root
elif operation == "sqrt":
    a = float(input("Enter number: "))
    if a < 0:
        print("Error: Cannot take square root of a negative number.")
    else:
        print("Result:", math.sqrt(a))

# Log base 10
elif operation == "log10":
    a = float(input("Enter number: "))
    if a <= 0:
        print("Error: Log undefined for zero or negative numbers.")
    else:
        print("Result:", math.log10(a))

else:
    print("Invalid operation. Please choose a valid option.")
```

Recursion is elegant, but not always practical.
Recursion is a bad idea when:
- the depth becomes too large
- the problem has overlapping subproblems (like Fibonacci)
- an iterative solution is simpler and faster
- Python’s recursion limit (≈1000) will be exceeded
- memory usage becomes too high
Examples where recursion is not ideal:
- Fibonacci (naive)
- Factorial for huge numbers
- Deep tree traversal without tail recursion
- Large linked list processing
Examples where recursion is ideal:
- DFS on graphs
- Tree traversal
- Divide‑and‑conquer algorithms (merge sort, quicksort)
- Backtracking (Sudoku, permutations)


```python
#71. Write recursive function for Fibonacci.

n= int(input("enter a number: "))

def fib(n):                                          #if n<=1 return n is known as base case.
    if n <= 1:                                #base case is a condition to stops the recursion.
        return n
    return fib(n-1) + fib(n-2)          #In fib(n):- here 'n' is a known as a  recursion depth
                                        #if n= 6 then it has 6 recursion depth
'''bigger the number of n then:- it will be more slow down
                                                                       - Stack oveflow
                                                                       - exponential time complexity
                                                                       '''
print(fib(n))


'''Once you understand recursive Fibonacci, you can understand:
- Dynamic programming
- Memoization
- Divide‑and‑conquer algorithms
- Tree recursion
- Backtracking
- Graph traversal (DFS)
'''
```

```python
#94. Check balanced parentheses.
expr = input("Enter an expression: ")
def is_balanced(expr):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for ch in expr:
        if ch in "({[":
            stack.append(ch)
        elif ch in ")}]":
            if not stack or stack.pop() != pairs[ch]:
                return False

    return len(stack) == 0

if is_balanced(expr):
    print("Balanced")
else:
    print("Not Balanced")
```

```python
#97. Find longest consecutive sequence in list.
lst = [100, 4, 200, 1, 5, 2,4,5,6]
def longest_consecutive(nums):
    if not nums:
        return 0

    num_set = set(nums)
    longest = 0

    for n in num_set:
        # Check if n is the start of a sequence
        if n - 1 not in num_set:
            length = 1
            current = n

            # Count forward
            while current + 1 in num_set:
                current += 1
                length += 1

            longest = max(longest, length)

    return longest


print("Longest consecutive sequence length:", longest_consecutive(lst))
```

```python
#99. Find majority element in list.
lst = [1,2,3,4,5,6,6,7,8,9,2,3,7,2]
def majority_element(lst):
  if not lst:
    return None
  count = 1
  candidate = lst[0]
  for i in range(1, len(lst)):
    if lst[i] == candidate:
      count += 1
    else:
      count -= 1
      if count == 0:
        candidate = lst[i]
        count = 1
  return candidate
print(majority_element(lst))
```

```python
#Convert a NumPy Array to a String
a1= np.array([1, 2, 3])
print(a1,type(a1))
s = a1.astype(str)
print('\nString: ',s,type(s))     #astype(str) does not give you one big string.It gives you an array of strings.
print(s.dtype)
"""
Notes:
output:- U21   where U means Unicode String. 21 is max space for the character.
my character space is currently 1 so it should show me U1 but somewhere in the past I have
given 21 characters to the a1 so even after I overwrite the a1 character space can't be resize automatically.
but it can be resize manually by using tolist
"""
print('To reset the dtype to minimal size')
s = np.array(s.tolist(), dtype=str)
print(s.dtype)
#converting String to Integer
print('\nConvert string to int')
i= s.astype(int)
print('Integer: ',i)             #arrays of int.
print(i.dtype,':- This says we converted to integer')

print('\nlets convert this integer to float')
f=i.astype(float)
print('Float: ',f)
print(f.dtype,':- This says we converted to float')

print('\nlets convert this float to integer')
i=f.astype(int)
print('Integer: ',i)
print(i.dtype,':- This says we converted to integer')
```

Stacking Arrays

```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

c = np.stack([a, b])
print(c)

d= np.stack([a, b],axis=1)
print('\n',d)

e= np.vstack([a, b])
print('\nvstack',e)

f= np.hstack([a, b])
print('\nhstack',f)

g= np.dstack([a, b])
print('\ndstack',g)

h= np.concatenate([a, b])
print('\nconcatenate',h)



```
